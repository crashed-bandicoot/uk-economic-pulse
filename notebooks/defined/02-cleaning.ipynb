# Cleaning the data

```python
import pandas as pd
```

```python
absolute_values = pd.read_csv('/Users/ornettematthews/Documents/GitHub/uk-economic-pulse/data/raw/absolute_values.csv', index_col=0, parse_dates=True)
```

```python
# --- Cleaning Checks ---

# 1. Shape and dtypes
print("Shape:", absolute_values.shape)
print()
print("Dtypes:")
print(absolute_values.dtypes)
print()

# 2. Missing values per column
missing = absolute_values.isna().sum()
missing_pct = (missing / len(absolute_values) * 100).round(1)
missing_report = pd.DataFrame({"missing": missing, "pct": missing_pct})
print("Missing values:")
print(missing_report[missing_report["missing"] > 0].to_string())
print()

# 3. Date index integrity — check for gaps in the monthly sequence
expected = pd.date_range(start=absolute_values.index.min(), end=absolute_values.index.max(), freq="ME")
missing_months = expected.difference(absolute_values.index)
print(f"Index gaps: {len(missing_months)} missing month(s)")
if len(missing_months):
    print(missing_months)
print()

# 4. Duplicate index entries
dupes = absolute_values.index.duplicated().sum()
print(f"Duplicate index entries: {dupes}")
print()

# 5. Columns with all-NaN (dead series)
all_null = absolute_values.columns[absolute_values.isna().all()].tolist()
print(f"All-NaN columns: {all_null if all_null else 'none'}")
print()

# 6. Date range per column (first and last valid observation)
print("Date coverage per column:")
coverage = pd.DataFrame({
    "first_valid": absolute_values.apply(lambda c: c.first_valid_index()),
    "last_valid":  absolute_values.apply(lambda c: c.last_valid_index()),
})
print(coverage.to_string())
```
